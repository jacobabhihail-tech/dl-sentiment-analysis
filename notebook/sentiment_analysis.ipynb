```python
import pandas as  pd
```

```python
data = {
    "review":[
        "I LOVE this movie!!!",
        "Worst film ever... waste of time",
        "It was okay, nothing special",
        "Amazing acting and great story",
        "Terrible movie, I hated it"
    ],
    "sentiment":[
        "positive",
        "negative",
        "neutral",
        "positive",
        "negative"
    ]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
import re
```

```python
def clean_text(text):

    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = re.sub(r'\s+', ' ', text)

    return text
```

```python
clean_text("OMG!!! This is BAD 123 😡")
```

```python
df['clean_review'] = df['review'].apply(clean_text)
```
